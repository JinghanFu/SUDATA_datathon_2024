```python
import pandas as pd
# Code to load the CSV file
file_path = '/kaggle/input/new-new-new/input_df(4).csv'
input_df = pd.read_csv(file_path)
input_df
```

# Only select day 6 open value and remove NA


```python
df_cleaned = input_df.drop(columns=['future_open_day_1','future_open_day_2', 'future_open_day_3', 'future_open_day_4','future_open_day_5','future_open_day_6'])

df_cleaned = df_cleaned.dropna()
```

```python
df_no_sentiment = df_cleaned.drop(columns=['sentiment_day_1','sentiment_day_2', 'sentiment_day_3', 'sentiment_day_4', 'sentiment_day_5'])
df_no_sentiment
```

# XGB


```python
import pandas as pd
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

def run_cross_val(df, model, n_splits=5, random_state=20241004, target_column='future_open_day_7'):
    df['target'] = df[target_column].apply(lambda x: 1 if x > 0 else 0)

    features = df.drop(columns=[target_column, 'target'])
    target = df['target']

    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    cross_val_scores = cross_val_score(model, features, target, cv=kf, scoring='accuracy')

   
    model.fit(features, target)

    if hasattr(model, 'feature_importances_'):
        feature_importances = model.feature_importances_
    else:
        feature_importances = None

    return cross_val_scores, feature_importances


# Initialize a dictionary to store results
results = {}

models = {
    'xgb': xgb.XGBClassifier(booster='gbtree', n_estimators=500, objective='binary:logistic', max_depth=3, learning_rate=0.01),
    'random_forest': RandomForestClassifier(n_estimators=500, max_depth=5, random_state=42),
    'logistic_regression': LogisticRegression(max_iter=1000, solver='lbfgs'),
    'gradient_boosting': GradientBoostingClassifier(n_estimators=500, learning_rate=0.01, max_depth=3)
}

# Run cross-validation for each model on df_cleaned (with sentiment)
for model_name, model in models.items():
    print(f"Running {model_name} on df_cleaned...")
    scores, importances = run_cross_val(df_cleaned, model)
    results[f'{model_name}_sentiment'] = scores
    print(f"{model_name} (With sentiment) - Mean CV Accuracy: {scores.mean():.4f}")

# Run cross-validation for each model on df_no_sentiment (without sentiment)
for model_name, model in models.items():
    print(f"Running {model_name} on df_no_sentiment...")
    scores, importances = run_cross_val(df_no_sentiment, model)
    results[f'{model_name}_no_sentiment'] = scores
    print(f"{model_name} (No sentiment) - Mean CV Accuracy: {scores.mean():.4f}")

# Print final results
print("\nResults Summary:")
for model_name, scores in results.items():
    print(f"{model_name} - Mean CV Accuracy: {scores.mean():.4f}")
```

```python

def run_powerful_lstm_cross_val(features, target, n_splits=5, random_state=20241004):
    # Step 1: Standardize the feature data
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Reshape the data into 3D array (samples, timesteps, features)
    X = np.reshape(scaled_features, (scaled_features.shape[0], 1, scaled_features.shape[1]))
    y = target.values

    kf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    fold_accuracies = []

    for train_idx, test_idx in kf.split(X, y):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        model = Sequential()
        model.add(LSTM(128, activation='relu', return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
        model.add(Dropout(0.3))  # Add dropout to prevent overfitting

        model.add(LSTM(64, activation='relu', return_sequences=False))  # A second LSTM layer
        model.add(Dropout(0.3))  # Add dropout for regularization

        model.add(Dense(32, activation='relu'))  # Fully connected layer
        model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification


        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    
        model.fit(X_train, y_train, epochs=5, batch_size=16, verbose=1, validation_data=(X_test, y_test))

        y_pred_prob = model.predict(X_test)
        y_pred = np.where(y_pred_prob > 0.5, 1, 0)

        fold_accuracy = accuracy_score(y_test, y_pred)
        fold_accuracies.append(fold_accuracy)
        print(f'Fold Accuracy: {fold_accuracy:.4f}')

    # Step 9: Print the mean accuracy across all folds
    mean_accuracy = np.mean(fold_accuracies)
    print(f'Mean Cross-Validation Accuracy: {mean_accuracy:.4f}')

    return mean_accuracy, fold_accuracies


target = df_cleaned['future_open_day_7'].apply(lambda x: 1 if x > 0 else 0)
features_with_sentiment = df_cleaned.drop(columns=['future_open_day_7', 'target'])

target_no_sentiment = df_no_sentiment['future_open_day_7'].apply(lambda x: 1 if x > 0 else 0)
features_no_sentiment = df_no_sentiment.drop(columns=['future_open_day_7', 'target'])

print("Results for dataset with sentiment features:")
mean_accuracy_with_sentiment, fold_accuracies_with_sentiment = run_powerful_lstm_cross_val(features_with_sentiment, target)
results['lstm_sentiment']= fold_accuracies_with_sentiment

print("\nResults for dataset without sentiment features:")
mean_accuracy_no_sentiment, fold_accuracies_no_sentiment = run_powerful_lstm_cross_val(features_no_sentiment, target_no_sentiment)
results['lstm_no_sentiment']= fold_accuracies_no_sentiment
```
